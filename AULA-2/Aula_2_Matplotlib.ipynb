Qual o TOP10 estados com mais quantidade de Registro de reflorastamento?

```python
# vsualizando gráfico com top10
x_bar = df['Estado sigla'].value_counts().head(10).index
y_bar = df['Estado sigla'].value_counts().head(10).values
```

```python
fig, ax = plt.subplots()
ax.bar(x_bar,y_bar)
for barra in ax.patches:
  ax.text(barra.get_x() + barra.get_width()/2,
          barra.get_height() + 25,
          barra.get_height(),
          ha='center')
```

Tranformando em uma função TOPN

```python
# função para topn
def top_n (n):
  x_bar = df['Estado sigla'].value_counts().head(n).index
  y_bar = df['Estado sigla'].value_counts().head(n).values
  fig, ax = plt.subplots()
  ax.bar(x_bar,y_bar)
  for barra in ax.patches:
    ax.text(barra.get_x() + barra.get_width()/2,
            barra.get_height() + 25,
            barra.get_height(),
            ha='center')
```

```python
top_n(5)
```

```python
top_n(7)
```

#### Gráfico de Linhas

```python
df.head(3)
```

Qual a soma total das áreas reflorestadas ao longo dos anos?

```python
# vários agrupamentos juntos
df.groupby(['Ano (data)']).agg(['sum','mean','count'])
```

```python
# agrupando o total reflorestado por ano
df.groupby(['Ano (data)']).sum()['Área (ha)']
```

```python
# identificando eixo x
df.groupby(['Ano (data)']).sum()['Área (ha)'].index
```

```python
# identificando eixo y
df.groupby(['Ano (data)']).sum()['Área (ha)'].values
```

```python
# visualizando o resultado em um gráfico de linha
x_plot = df.groupby(['Ano (data)']).sum()['Área (ha)'].index
y_plot = df.groupby(['Ano (data)']).sum()['Área (ha)'].values
plt.plot(x_plot,y_plot)
plt.show()
```

Qual a soma total das áreas reflorestadas ao longo dos anos observando as espécies?


```python
# agrupando os a soma por data e especie
df.groupby(['Ano (data)', 'Espécie florestal']).sum()\
  .reset_index()\
  .pivot('Ano (data)', 'Espécie florestal', 'Área (ha)')
```

```python
# transformando a consulta em um dataframe
df_especie = df.groupby(['Ano (data)', 'Espécie florestal']).sum()\
              .reset_index()\
              .pivot('Ano (data)', 'Espécie florestal', 'Área (ha)')\
              .reset_index()

df_especie
```

```python
# visualizando o resultado em um gráfico de linha
plt.plot('Ano (data)','Eucalipto', data=df_especie, label='Eucalipto')
plt.plot('Ano (data)','Outras espécies', data=df_especie, label='Outras espécies')
plt.plot('Ano (data)','Pinus', data=df_especie, label='Pinus')
plt.legend(ncol=3, loc = 'upper left')
plt.show()
```

```python
# visualizando o resultado em um gráfico de linha
plt.plot('Ano (data)','Eucalipto', data=df_especie, label='Eucalipto')
plt.plot('Ano (data)','Outras espécies', data=df_especie, label='Outras espécies')
plt.plot('Ano (data)','Pinus', data=df_especie, label='Pinus')
plt.plot(x_plot,y_plot, label='Total')
plt.legend(ncol=4, loc = 'upper left')
plt.show()
```
